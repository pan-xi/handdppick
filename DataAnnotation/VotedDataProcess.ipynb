```python
import matplotlib.pyplot as plt

# 示例数据（假设从文件中提取的每个模型的总得分）
model_scores = {
    'gpt4o': 1200,
    'deepseekv3': 900,
    'claude3.5-1022': 1100,
    'gemini-1.5-pro-latest': 800,
    'yi-lightning': 1000
}

# 计算每个模型得分的比例
total_score = sum(model_scores.values())
percentages = [score / total_score * 100 for score in model_scores.values()]
labels = list(model_scores.keys())

# 颜色设置
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

# 绘制环形图
fig, ax = plt.subplots(figsize=(8, 8))

# 设置每个环的半径（从内到外依次增加）
radii = [0.5, 0.7, 0.9, 1.1, 1.3]  # 从中心到外层的半径
ring_width = 0.2  # 每个环的宽度

# 绘制每个环
for i, (radius, percentage, label, color) in enumerate(zip(radii, percentages, labels, colors)):
    # 绘制环形图
    wedges, _ = ax.pie(
        [percentage, 100 - percentage],  # 当前环的比例和空白部分
        radius=radius,
        colors=[color, 'white'],  # 当前环的颜色和空白部分的颜色
        wedgeprops=dict(width=ring_width, edgecolor='w')
    )
    # 添加标签
    angle = wedges[0].theta2  # 获取当前环的结束角度
    x = radius * 0.8 * np.cos(np.deg2rad(angle))  # 标签的x坐标
    y = radius * 0.8 * np.sin(np.deg2rad(angle))  # 标签的y坐标
    ax.text(x, y, label, ha='center', va='center', fontsize=10, color='black')

# 设置标题
ax.set_title('Stacked Donut Chart of Model Scores (Inner to Outer)', fontsize=16)

# 显示图形
plt.show()
```
